```python
import pandas as pd

# Dosya isimlerini liste halinde girin
csv_dosyalari = ['E:/patch_crop/Breast1__he.csv', 'E:/patch_crop/Breast2__he.csv', 'E:/patch_crop/Breast3__ihc.csv', 'E:/patch_crop/LymphNode1__he.csv',"E:/patch_crop/Prostate1__he.csv"]

# Tüm dosyaları birleştirmek için boş bir DataFrame oluşturun
birlesik_df = pd.DataFrame()

# Her dosyayı okuyup birleştirin
for dosya in csv_dosyalari:
    df = pd.read_csv(dosya)
    birlesik_df = pd.concat([birlesik_df, df], ignore_index=True)

# Birleşik DataFrame'i yeni bir CSV dosyasına kaydedin
csv_path = "E:/patch_crop/birlesik_df.csv"
birlesik_df.to_csv(csv_path, index=False)
```

```python
import pandas as pd
import os

# Verilen DataFrame'i karıştırıp eğitim ve doğrulama setlerine ayıran fonksiyon
def split_data_frame(df, train_ratio):
    shuffled_df = df.sample(frac=1).reset_index(drop=True)
    train_size = int(len(df) * train_ratio)
    train_df = shuffled_df.iloc[:train_size]
    valid_df = shuffled_df.iloc[train_size:]
    return train_df, valid_df

# Belirtilen klasörden dosya yollarını alan fonksiyon
def get_files_from_folder(df, column_name):
    files = []
    for folder in df[column_name]:
        folder_path = os.path.normpath(folder)  # Dosya yolunu standartlaştır
        folder_path = folder_path.replace('\\', '/')  # Tüm ters eğik çizgileri normal eğik çizgi ile değiştir
        files.extend([os.path.join(folder_path, f).replace('\\', '/') for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
    return files

# CSV dosyalarınızın yollarını bir listeye koyun
csv_files = ['E:/patch_crop/birlesik_df.csv']

# Tüm eğitim ve doğrulama setlerini saklayacak listeler
all_x_train, all_y_train, all_x_valid, all_y_valid = [], [], [], []

# Her bir CSV dosyası için işlemleri yap
for csv_file in csv_files:
    df = pd.read_csv(csv_file)  # DataFrame'i yükle
    train_df, valid_df = split_data_frame(df, 0.875)  # DataFrame'i ayır

    # Dosya yollarını al
    x_train = get_files_from_folder(train_df, 'Image Folder')
    y_train = get_files_from_folder(train_df, 'Mask Folder')
    x_valid = get_files_from_folder(valid_df, 'Image Folder')
    y_valid = get_files_from_folder(valid_df, 'Mask Folder')

    # Elde edilen yolları genel listelere ekle
    all_x_train.extend(x_train)
    all_y_train.extend(y_train)
    all_x_valid.extend(x_valid)
    all_y_valid.extend(y_valid)

# Şimdi all_x_train, all_y_train, all_x_valid, all_y_valid tüm CSV dosyalarından elde edilen yolları içeriyor.
```

```python
from skimage.io import imread
from skimage.transform import resize
import numpy as np
from joblib import Parallel, delayed

# Yeniden boyutlandırılacak boyutlar
IMG_WIDTH = 128
IMG_HEIGHT = 128
IMG_CHANNELS = 3

# Eğitim ve doğrulama setlerinizin yolları
train_image_paths = all_x_train
train_mask_paths = all_y_train
valid_image_paths = all_x_valid
valid_mask_paths = all_y_valid
test_image_paths = x_test
test_mask_paths = y_test


# Boş dizilerin oluşturulması
X_train_resized = np.zeros((len(train_image_paths), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
Y_train_resized = np.zeros((len(train_mask_paths), IMG_HEIGHT, IMG_WIDTH, 1), dtype=bool)
X_valid_resized = np.zeros((len(valid_image_paths), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
Y_valid_resized = np.zeros((len(valid_mask_paths), IMG_HEIGHT, IMG_WIDTH, 1), dtype=bool)
X_test_resized = np.zeros((len(test_image_paths), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
Y_test_resized = np.zeros((len(test_mask_paths), IMG_HEIGHT, IMG_WIDTH, 1), dtype=bool)

def load_resize_image_mask(image_path, mask_path):
    img = imread(image_path)[:,:,:IMG_CHANNELS]
    img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)

    mask = imread(mask_path)
    mask = np.expand_dims(resize(mask, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True), axis=-1)

    return img, mask

# Parallelize the loading and resizing process for the training set
result_train = Parallel(n_jobs=-1)(delayed(load_resize_image_mask)(image_path, mask_path)
                                   for image_path, mask_path in zip(train_image_paths, train_mask_paths))

# Unpack the result for the training set
X_train_resized, Y_train_resized = zip(*result_train)
X_train_resized = np.array(X_train_resized)
Y_train_resized = np.array(Y_train_resized)

# Parallelize the loading and resizing process for the validation set
result_valid = Parallel(n_jobs=-1)(delayed(load_resize_image_mask)(image_path, mask_path)
                                   for image_path, mask_path in zip(valid_image_paths, valid_mask_paths))

# Unpack the result for the validation set
X_valid_resized, Y_valid_resized = zip(*result_valid)
X_valid_resized = np.array(X_valid_resized)
Y_valid_resized = np.array(Y_valid_resized)

result_test = Parallel(n_jobs=-1)(delayed(load_resize_image_mask)(image_path, mask_path)
                                  for image_path, mask_path in zip(test_image_paths, test_mask_paths))

# Unpack the result for the test set
X_test_resized, Y_test_resized = zip(*result_test)
X_test_resized = np.array(X_test_resized)
Y_test_resized = np.array(Y_test_resized)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D

# Basit bir CNN tabanlı segmentasyon modeli oluşturma
def simple_segmentation_model(input_size=(128, 128, 3)):
    inputs = Input(input_size)
    
    # Encoder (Daraltıcı Yol)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
    
    # Bottleneck
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    
    # Decoder (Genişletici Yol)
    up1 = UpSampling2D(size=(2, 2))(conv2)
    up1 = Conv2D(64, 2, activation='relu', padding='same')(up1)
    
    # Çıkış katmanı
    output = Conv2D(1, 1, activation='sigmoid')(up1)
    
    model = tf.keras.Model(inputs=inputs, outputs=output)
    
    return model

# Modeli oluşturma
input_size = (128, 128, 3)  # Giriş boyutları (örnek olarak 128x128 piksel ve 3 kanal)
model = simple_segmentation_model(input_size)

# Model özetini görüntüleme
model.summary()
```
